### Czarno biały filtr

```python
im2[:,:,1]=im2[:,:,0]
im2[:,:,2]=im2[:,:,0]
 
plt.imshow(im2) 
```

```python
sum_arrey = im2.sum(axis=2,keepdims=True)
sum_arrey = sum_arrey/np.amax(sum_arrey)#100

sum_arrey.shape
data = sum_arrey*np.ones(3)
plt.imshow(data)
```
